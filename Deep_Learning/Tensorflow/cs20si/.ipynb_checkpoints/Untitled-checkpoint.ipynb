```python
import tensorflow as tf
```

```python
a = tf.constant(2)
b = tf.constant(3)

x = tf.add(a,b)

with tf.Session() as sess:
    writer = tf.summary.FileWriter('./graphs', sess.graph) # used to store the graph
    # here we are telling it to store the graph at the specified location
    print(sess.run(x))
    
writer.close() # important to close the writer
```

After running this, open the terminal to your project directory.
Type
**tensorboard --logdir="./graphs" --port 6006**. This provides a web address for tensorboard. In the graph section, we can actually visualise our graph.

In the graph we observe , that a and b have been given names 'constant' and 'constant_1'. These are the default names of the graph. We can change them. For this..

```python
import tensorflow as tf

a = tf.constant(2, name="a")
b = tf.constant(3, name="b") # gives the name in the tensorboard

x = tf.add(a,b)

with tf.Session() as sess:
    writer = tf.summary.FileWriter('./graphs', sess.graph)
    print(sess.run(x))
    
writer.close()
```

** Constants are stored in the graph definition. So constants are not so good **. To see the graph definition.....

```python
import tensorflow as tf

my_const = tf.constant([1,2], name="my_constant")
with tf.Session() as sess:
    print(sess.graph.as_graph_def())

# we will see the value of my_const stored in graph definition.
```

## VARIABLES


```python
a= tf.Variable(initial_value=2, name='a')
# The initial value defines the
# type and shape of the variable. After construction, the type and shape of
# the variable are fixed. Its value can be changed using the assign operator

b = tf.Variable([2,3], name='b')

c = tf.Variable(tf.zeros([2,2]))
```

## Why Variable with capital V

* Constant is an operation and variable is a class. So many functions can be called on Variables

* A bunch of operations come pre packaged with a Variable.


```python
x = tf.Variable(2)

x.initializer # init op
x.value() # read op
x.assign() # assigns a value to x
x.assign_add() # add the value in parameter to x
```

Since all these are ops they need to be called in sess.run()

## ALL VARIABLES NEED TO BE INITIALIZED IN TENSORFLOW



```python
# to initialise all the variables
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
```

To initialize a subset of Variables

```python
init_ab = tf.variables_initializer([a,b], name='init_ab')
with tf.Session() as sess:
    sess.run(init_ab)
```

To initialise a single variable. This initializes the Variable with the initial value

```python
temp = tf.Variable(0)
with tf.Session() as sess:
    sess.run(temp.initializer)
```

## To print a Variable use eval()

Calling this method will execute all preceding operations that
produce the inputs needed for the operation that produces this
tensor.


```python
x = tf.Variable(1)
with tf.Session() as sess:
    sess.run(x.initializer)
    print(x.eval())
    print("Done")
```

## tf.Variable.assign() -> Assigns a value to a Variable

Since tf.Variable.assign() is an op, it needs to be called in the sess.

```python
x = tf.Variable(10)
assign = x.assign(100)
with tf.Session() as sess:
    sess.run(assign)
    print(x.eval())
```

## Every session maintains its own copy of the Variable

```python
W = tf.Variable(10)
sess1 = tf.Session()
sess2 = tf.Session()

sess1.run(W.initializer)
sess2.run(W.initializer)

print(sess1.run(W.assign_add(10)))
print(sess2.run(W.assign_sub(2)))

sess1.close()
sess2.close()
```

## If we want to initialize a Variable with another Variable's initialized value..

We need to make sure the order of initializing the variables. For that use 
** tf.Variable.initialized_value()**

```python
a = tf.Variable(1)
b = tf.Variable(2* a.initialized_value())
```

# PlaceHolders

We can assemble the graph first without knowing the actual data. 
Analogous to f(x) = x**2 . We can define it, without knowing the actual value of x. 

Useful as this can allow us to put in the value later.

```python
a = tf.placeholder(tf.float32, shape=[3])
b = tf.constant([1,1,1],tf.float32)

c = tf.add(a,b) # we can use a as if we already know its value
```

** Feed the values of the placeholder using a dictionary**

```python
with tf.Session() as sess:
    print(sess.run(c, feed_dict={a:[1,2,3]})) # a not 'a' is the key
```

shape=None means that the placeholder with accept any tensor of any shape
